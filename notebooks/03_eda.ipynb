# 3. Exploratory Data Analysis

## Objective

The goal of this notebook is to explore relationships between climate variables and crop yield through visualization and analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/climate_change_impact_on_agriculture_2024.csv")
```

## 3.1 Distribution of Crop Yield

This section examines the distribution of the target variable to understand its overall shape and variability.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Crop_Yield_MT_per_HA"], bins=30, kde=True)
plt.title("Distribution of Crop Yield")
plt.xlabel("Crop Yield (MT per HA)")
plt.ylabel("Frequency")
plt.show()
```

### Interpretation

- The distribution of crop yield appears moderately right-skewed.
- Most values are concentrated between approximately 1.5 and 3.5 MT per HA.
- The distribution shows a gradual decline toward higher values, indicating fewer high-yield observations.
- There are no extreme outliers, and the range appears bounded between approximately 0.5 and 5.

Overall, the distribution suggests a controlled and relatively smooth spread of values, which may indicate a well-structured or possibly simulated dataset.

## 3.2 Temperature vs Crop Yield

This section explores the relationship between temperature and crop yield.

```python
sns.scatterplot(
    x="Average_Temperature_C",
    y="Crop_Yield_MT_per_HA",
    data=df,
    alpha=0.3
)
```

### Interpretation

- The relationship between temperature and crop yield does not appear to follow a clear linear pattern.
- Instead, the data forms distinct rectangular clusters across temperature ranges.
- These structured patterns are unlikely to occur in purely observational real-world data.
- This suggests that the dataset may be artificially generated or discretized.

As a result, while temperature may appear to influence yield within certain ranges, the relationship should be interpreted with caution.

## 3.3 Precipitation vs Crop Yield

This section explores the relationship between precipitation and crop yield.

```python
sns.scatterplot(
    x="Total_Precipitation_mm",
    y="Crop_Yield_MT_per_HA",
    data=df,
    alpha=0.3
)
```

### Interpretation

- There is no strong visible relationship between precipitation and crop yield.
- Data points are widely dispersed across all precipitation levels.
- The absence of a clear trend suggests weak or no direct association in this dataset.

However, given the structured nature of the dataset, this lack of relationship may reflect how the data was generated rather than real-world dynamics.

## 3.4 Crop Type vs Crop Yield

This section examines how crop yield varies across different crop types.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Crop_Type", y="Crop_Yield_MT_per_HA", data=df)
plt.title("Crop Yield Distribution by Crop Type")
plt.xlabel("Crop Type")
plt.ylabel("Crop Yield (MT per HA)")
plt.xticks(rotation=45)
plt.show()
```

### Interpretation

- Crop yield appears to be very similar across different crop types.
- The median, spread, and overall distribution are nearly identical for all crops.
- This is unusual, as real-world agricultural data typically shows variation between crop types.

These patterns suggest that crop type may not be a strong differentiating factor in this dataset.  
Alternatively, the uniformity across groups indicates that the data may be artificially generated or simplified.

Therefore, the observed similarities should be interpreted with caution.

## 3.5 Region vs Crop Yield

This section examines how crop yield varies across regions.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Region", y="Crop_Yield_MT_per_HA", data=df)
plt.title("Crop Yield Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Crop Yield")
plt.xticks(rotation=45)
plt.show()
```

### Interpretation

- Crop yield appears to be highly consistent across different regions.
- The median values and overall distribution are nearly identical for all regions.
- This lack of variation is unexpected, as real-world agricultural data typically shows regional differences.

These patterns suggest that regional factors may not be meaningfully represented in this dataset.  
Alternatively, the uniform distributions indicate that the data may be artificially generated or simplified.

Therefore, the results should be interpreted with caution.

## 3.6 Correlation Analysis

This section examines relationships between numerical variables using correlation.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### Interpretation

- Most variables show very weak or near-zero correlation with crop yield.
- Temperature has a modest positive correlation with yield (approximately 0.26).
- Precipitation and other environmental variables show little to no correlation.

- A strong positive correlation is observed between crop yield and economic impact (approximately 0.73).
- This unusually high correlation suggests that economic impact may be directly derived from yield rather than independently measured.

Overall, the correlation structure appears overly simplified, indicating that the dataset may be artificially generated or structured.  
Therefore, the relationships observed should be interpreted with caution.

## 3.7 Irrigation Access vs Crop Yield

This section explores the relationship between irrigation access and crop yield.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Irrigation_Access_%",
    y="Crop_Yield_MT_per_HA",
    data=df,
    alpha=0.3
)
plt.title("Irrigation Access vs Crop Yield")
plt.xlabel("Irrigation Access (%)")
plt.ylabel("Crop Yield")
plt.show()
```

### Interpretation

- There is no clear relationship between irrigation access and crop yield.
- Data points are evenly distributed across all levels of irrigation.
- The absence of a visible trend suggests that irrigation does not significantly influence yield in this dataset.

This lack of association further supports the idea that the dataset may be artificially structured.

## 3.8 Fertilizer Use vs Crop Yield

This section explores the relationship between fertilizer use and crop yield.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Fertilizer_Use_KG_per_HA",
    y="Crop_Yield_MT_per_HA",
    data=df,
    alpha=0.3
)
plt.title("Fertilizer Use vs Crop Yield")
plt.xlabel("Fertilizer Use (kg per HA)")
plt.ylabel("Crop Yield")
plt.show()
```

### Interpretation

- No clear relationship is observed between fertilizer use and crop yield.
- The distribution of points is uniform across all fertilizer levels.
- This suggests that fertilizer input does not strongly affect yield within this dataset.

As with other variables, this pattern indicates a simplified or synthetic data structure.

## 3.9 Key Findings

- The distribution of crop yield is moderately right-skewed, with most values concentrated between approximately 1.5 and 3.5 MT per HA.

- Climate variables such as temperature show only weak to moderate relationships with crop yield, while precipitation and extreme weather events show little to no clear association.

- Crop type and region do not significantly differentiate yield distributions, as their patterns are nearly identical across groups.

- Agricultural inputs such as irrigation access and fertilizer use also show no strong relationship with crop yield, with data points evenly distributed across all values.

- A strong positive correlation is observed between crop yield and economic impact, suggesting that economic impact may be directly derived from yield rather than independently measured.

- Overall, the dataset exhibits highly uniform and structured patterns across multiple variables, which is unlikely in real-world agricultural data.

These findings suggest that the dataset may be artificially generated or simplified, and therefore the observed relationships should be interpreted with caution.
